```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# Output File (CSV)
# output_data_file = "output_data/cities.csv"

```

```python
# Take in all of our census data and read it into pandas
census_data_2015_clean = os.path.join('Resources','Census Data',"census_data_2015_clean.csv")
census_data_2016_clean = os.path.join('Resources','Census Data',"census_data_2016_clean.csv")
census_data_2017_clean = os.path.join('Resources','Census Data',"census_data_2017_clean.csv")

census_data_2015_clean_df=pd.read_csv(census_data_2015_clean)
census_data_2016_clean_df=pd.read_csv(census_data_2016_clean)
census_data_2017_clean_df=pd.read_csv(census_data_2017_clean)
```

```python
# Merge 2015 and 2016 data into new DF
combined_census_df = pd.merge(census_data_2015_clean_df, census_data_2016_clean_df, how="outer", on="Zipcode")
combined_census_df.head()
```

```python
# Rename columns to include census year name
combined_census_df = combined_census_df.rename(columns={"Population_x":"2015 Population",
                                                        "Median Age_x":"2015 Median Age",
                                                        "Household Income_x":"2015 Household Income",
                                                        "Per Capita Income_x":"2015 Per Capita Income",
                                                        "Population_y":"2016 Population",
                                                        "Median Age_y":"2016 Median Age",
                                                        "Household Income_y":"2016 Household Income",
                                                        "Per Capita Income_y":"2016 Per Capita Income"})
combined_census_df.head()
```

```python
# Add 2017 data to new DF
combined_census_df = pd.merge(combined_census_df, census_data_2017_clean_df, how="outer", on="Zipcode")
combined_census_df.head()
```

```python
# Rename added 2017 columns
combined_census_df = combined_census_df.rename(columns={"Population":"2017 Population",
                                                        "Median Age":"2017 Median Age",
                                                        "Household Income":"2017 Household Income",
                                                        "Per Capita Income":"2017 Per Capita Income"})
combined_census_df.head()
```

```python
# Take in Avocado Region zip codes and read into panda
city_zipcodes_avocado_regions = os.path.join("data","city_zipcodes_avocado_regions_new.csv")

city_zipcodes_avocado_regions_df=pd.read_csv(city_zipcodes_avocado_regions)

city_zipcodes_avocado_regions_df.head()
```

```python
# Merge 2015 census data and Avocado Regions data into new DF
combined_census_and_regions_2015_df = pd.merge(city_zipcodes_avocado_regions_df, census_data_2015_clean_df, how='inner', on='Zipcode')
combined_census_and_regions_2015_df.head()
```
