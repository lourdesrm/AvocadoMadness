```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt

# import CSV's
region_type_file = 'Resources/dfs/Price by Type by Region.csv'
summary_vol_file = 'Resources/dfs/Average Price and Volume Consumed by Region.csv'
avocado_summary_file = 'Resources/dfs/Average Price and Volume Consumption Summary.csv'
region_type_file = 'Resources/dfs/Price by Type by Region.csv'
avg_df1_file = 'Resources/dfs/Average Consumption by Year by Type.csv'
total_avo_volume_file = 'Resources/dfs/Convetional vs Organic Consumption.csv'

# read CSV files into new dfs
region_type_df = pd.read_csv(region_type_file)
summary_vol_price = pd.read_csv(summary_vol_file)
avocado_summary = pd.read_csv(avocado_summary_file)
region_type_df = pd.read_csv(region_type_file)
avg_df1 = pd.read_csv(avg_df1_file).drop("Unnamed: 0",1)
total_avo_volume = pd.read_csv(total_avo_volume_file).drop("Unnamed: 0",1)
```

```python
labels = region_type_df.reset_index(drop=False)
labels_name = labels['region'].unique()

min_price_conventional = labels[labels['type']=='conventional']['Minimum Price']

min_price_organic = labels[labels['type']=='organic']['Minimum Price']

max_price_conventional = labels[labels['type']=='conventional']['Maximum Price']

max_price_organic = labels[labels['type']=='organic']['Maximum Price']
```

```python
# the label locations
x = np.arange(len(labels_name))  
# the width of the bars
width = 0.35  
# setting subplot size
fig, ax = plt.subplots(figsize=(20,5))
# setting bar plots 
rects1 = ax.bar(x - width/2, min_price_conventional, width, label='Min Conventional Price')
rects2 = ax.bar(x + width/2, max_price_conventional, width, label='Max Conventional Price')
# Setting Y axis label
ax.set_ylabel('Price')
# Setting Plot Title
ax.set_title('Min/Max Price of Convetional Avocaodos 2015-2017')
# Setting X ticks
ax.set_xticks(x)
# Setting Y ticks with Y limit and intervals
ax.set_yticks(np.arange(0, max_price_conventional.max() + 0.1, step=0.25))
# Setting X label ticks
ax.set_xticklabels(labels_name, rotation = 45, ha="right")
# Setting up legend
ax.legend()
# Setting X limit
plt.xlim(-1.0,len(x))
# Setting tick parameters and width
ax.tick_params(which='minor', width=1.00)
# Cleaning up layout
fig.tight_layout()
# Saving as PNG for presentation
plt.savefig("Resources/screenshots/MinMaxConventionalPrices.png")
# Displaying plot
plt.show()
```

```python
# Create Plot Labels
labels = summary_vol_price
total_volume  = summary_vol_price["Total Volume(in millions)"]
# the label locations
x = np.arange(len(labels_name))  
# the width of the bars
width = 0.35  
# setting subplot size
fig, ax = plt.subplots(figsize=(20,5))
# setting bar plots
rects1 = ax.bar(x - width, total_volume, width, label='Avocados (in millions)')
#setting Y axis label to Avocado Consumption (in millions)
ax.set_ylabel('Avocados Consumed (in millions)')
# Set plot title
ax.set_title('Avocado Consumption by Region')
# Set the X axis ticks
ax.set_xticks(x)
#set Y axis ticks and intervals
ax.set_yticks(np.arange(0, 1000, step = 100))
# format x axis tick labels to make sure all labels are easy to digest
ax.set_xticklabels(labels_name, rotation = 45, ha="right")
# setting x axis limit
plt.xlim(-1.0,len(x))
# set tick parameters and width
ax.tick_params(which='minor', width=1.00)
# clean up layout
fig.tight_layout()
# save the plot as a PNG file for presentation
plt.savefig("Resources/screenshots/ConsumptionbyRegion.png")
# display plot
plt.show()
```

```python
# Calling min and max price for bars in plot
labels = avocado_summary
min_price = labels['Minimum Price']
max_price = labels['Maximum Price']
# the label locations
x = np.arange(len(labels_name))  
# the width of the bars
width = 0.35  
# setting subplot size
fig, ax = plt.subplots(figsize=(20,5))
# creating bar plots for min and max price from variables above
rects1 = ax.bar(x - width/2, min_price, width, label='Min Price')
rects2 = ax.bar(x + width/2, max_price, width, label='Max Price')
# setting Y axis label
ax.set_ylabel('Price')
# Setting plot title
ax.set_title('Min/Max Price of Avocaodos 2015-2017')
# setting x axis ticks
ax.set_xticks(x)
# setting y axis ticks while using max price to determine y limit and intervals
ax.set_yticks(np.arange(0, max_price.max() + 0.1, step=0.25))
# format x axis labels to be able to easily digest
ax.set_xticklabels(labels_name, rotation = 45, ha="right")
# show plot legend
ax.legend()
# set x axis limit
plt.xlim(-1.0,len(x))
# set tick parameters and width
ax.tick_params(which='minor', width=1.00)
# clean up plot
fig.tight_layout()
# save file as PNG for presentation
plt.savefig("Resources/screenshots/MinMaxPrices.png")
# show figure 
plt.show()
```
