## AVOCADO CONSUMPTION AND PRICING PLOTS

```python
# import dependencies
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt

# import CSV's
region_type_file = 'Resources/dfs/Price by Type by Region.csv'
summary_vol_file = 'Resources/dfs/Average Price and Volume Consumed by Region.csv'
avocado_summary_file = 'Resources/dfs/Average Price and Volume Consumption Summary.csv'
region_type_file = 'Resources/dfs/Price by Type by Region.csv'
avg_df1_file = 'Resources/dfs/Average Consumption by Year by Type.csv'
total_avo_volume_file = 'Resources/dfs/Convetional vs Organic Consumption.csv'
avg_inflated_df_file = 'Resources/dfs/Average Consumption by Year by Type.csv'

# read CSV files into new dfs
region_type_df = pd.read_csv(region_type_file)
summary_vol_price = pd.read_csv(summary_vol_file)
avocado_summary = pd.read_csv(avocado_summary_file)
region_type_df = pd.read_csv(region_type_file)
avg_df1 = pd.read_csv(avg_df1_file).set_index('Year')
total_avo_volume = pd.read_csv(total_avo_volume_file).set_index('Year')
avg_inflated_df = pd.read_csv(avg_inflated_df_file).set_index('Year')
```

### CREATING LABELS FOR PLOTTING

```python
# creating new df for labelling plots
labels = region_type_df.reset_index(drop=False)
# creating label names for each region
labels_name = labels['region'].unique()
# finding minimum price of conventional avocados
min_price_conventional = labels[labels['type']=='conventional']['Minimum Price']
# finding the minimum price for the organic avocados
min_price_organic = labels[labels['type']=='organic']['Minimum Price']
# finding the maximum price for conventional avocados
max_price_conventional = labels[labels['type']=='conventional']['Maximum Price']
# finding the maximum price for organic avocados
max_price_organic = labels[labels['type']=='organic']['Maximum Price']
```

### PLOTTING AVOCADO CONSUMPTION BY REGION

```python
# Create Plot Labels
labels = summary_vol_price
total_volume  = summary_vol_price["Total Volume(in millions)"]
# the label locations
x = np.arange(len(labels_name))  
# the width of the bars
width = 0.35  
# setting subplot size
fig, ax = plt.subplots(figsize=(20,5))
# setting bar plots to total volume
rects1 = ax.bar(x - width, total_volume, width, label='Avocados (in millions)')
#setting Y axis label to Avocado Consumption (in millions)
ax.set_ylabel('Avocados Consumed (in millions)')
# Set plot title
ax.set_title('Avocado Consumption by Region')
# Set the X axis ticks
ax.set_xticks(x)
#set Y axis ticks and intervals
ax.set_yticks(np.arange(0, 1000, step = 100))
# format x axis tick labels to make sure all labels are easy to digest
ax.set_xticklabels(labels_name, rotation = 45, ha="right")
# setting x axis limit
plt.xlim(-1.0,len(x))
# set tick parameters and width
ax.tick_params(which='minor', width=1.00)
# clean up layout
fig.tight_layout()
# save the plot as a PNG file for presentation
plt.savefig("Resources/screenshots/ConsumptionbyRegion.png")
# display plot
plt.show()
```

### PLOTTING MINIMUM AND MAXIMUM PRICE FOR ALL AVOCADOS 2015-2017

```python
# Calling min and max price for bars in plot
labels = avocado_summary
min_price = labels['Minimum Price']
max_price = labels['Maximum Price']
# the label locations
x = np.arange(len(labels_name))  
# the width of the bars
width = 0.35  
# setting subplot size
fig, ax = plt.subplots(figsize=(20,5))
# creating bar plots for min and max price from min and max price above
rects1 = ax.bar(x - width/2, min_price, width, label='Minimum Price')
rects2 = ax.bar(x + width/2, max_price, width, label='Maximum Price')
# setting Y axis label
ax.set_ylabel('Price')
# Setting plot title
ax.set_title('Min/Max Price of All Avocaodos 2015-2017')
# setting x axis ticks
ax.set_xticks(x)
# setting y axis ticks while using max price to determine y limit and intervals
ax.set_yticks(np.arange(0, max_price.max() + 0.1, step=0.25))
# format x axis labels to be able to easily digest
ax.set_xticklabels(labels_name, rotation = 45, ha="right")
# show plot legend
ax.legend()
# set x axis limit
plt.xlim(-1.0,len(x))
# set tick parameters and width
ax.tick_params(which='minor', width=1.00)
# clean up plot
fig.tight_layout()
# save file as PNG for presentation
plt.savefig("Resources/screenshots/MinMaxPrices.png")
# show figure 
plt.show()
```

### PLOTTING MINIMUM AND MAXIMUM PRICE FOR CONVENTIONAL AVOCADOS 2015-2017

```python
# the label locations
x = np.arange(len(labels_name))  
# the width of the bars
width = 0.35  
# setting subplot size
fig, ax = plt.subplots(figsize=(20,5))
# setting bar plots with min and max price for conventional avocados
rects1 = ax.bar(x - width/2, min_price_conventional, width, label='Minimum Price')
rects2 = ax.bar(x + width/2, max_price_conventional, width, label='Maximum Price')
# Setting Y axis label
ax.set_ylabel('Price')
# Setting Plot Title
ax.set_title('Min/Max Price of Convetional Avocaodos 2015-2017')
# Setting X ticks
ax.set_xticks(x)
# Setting Y ticks with Y limit and intervals
ax.set_yticks(np.arange(0, max_price_conventional.max() + 0.1, step=0.25))
# Setting X label ticks
ax.set_xticklabels(labels_name, rotation = 45, ha="right")
# Setting up legend
ax.legend()
# Setting X limit
plt.xlim(-1.0,len(x))
# Setting tick parameters and width
ax.tick_params(which='minor', width=1.00)
# Cleaning up layout
fig.tight_layout()
# Saving as PNG for presentation
plt.savefig("Resources/screenshots/MinMaxConventionalPrices.png")
# Displaying plot
plt.show()
```

### PLOTTING PRICE OF CONVENTIONAL AVOCADOS VS NATIONAL INFLATION

```python
# creating new df for conventional pricing by year
avg_df2 = avg_inflated_df
# dropping columns we do not want to plot (organic avocado pricing)
avg_df2 = avg_df2.drop('organic',1)
avg_df2 = avg_df2.drop('Organic Inflated',1)
# creating new plot for conventional pricing of avocados vs inflation
ax3 = avg_df2.plot()
# setting Y axis label
ax3.set_ylabel('Price')
# setting X axis label
ax3.set_xlabel('Year')
# setting X axis ticks
ax3.set_xticks([2015,2016,2017])
# setting plot title
ax3.set_title('Conventional Price of Avocados vs. Inflation Rate')
# display figure and save the plot as PNG for presentation
ax3.get_figure().savefig('Resources/screenshots/ConventionalPriceVsInflation.png')
# save df as CSV file for reference
export_csv = avg_df2.round(2).to_csv('Resources/dfs/ConventionalInflation.csv')
```

### PLOTTING PRICE OF ORGANIC AVOCADOS VS NATIONAL INFLATION

```python
# creating new data frame to drop conventional avocado prices
avg_df3 = avg_inflated_df
# dropping conventional avocado prices
avg_df3 = avg_df3.drop('conventional',1)
avg_df3 = avg_df3.drop('Conventional Inflated',1)
# plotting our df
ax4 = avg_df3.plot()
# setting Y axis label
ax4.set_ylabel('Price')
# set X Axis label
ax4.set_xlabel('Year')
# setting X axis ticks
ax4.set_xticks([2015,2016,2017])
# setting plot title
ax4.set_title('Organic Price of Avocados vs. Inflation Rate')
# show the plot and save as PNG for presentation
ax4.get_figure().savefig('Resources/screenshots/OrganicPriceVsInflation.png')
# save results dataframe
export_csv = avg_df3.round(2).to_csv('Resources/dfs/OrganicInflation.csv')
```
